```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import scipy as sp
url='https://raw.githubusercontent.com/Geoyi/Cleaning-Titanic-Data/master/titanic_original.csv'
titanic = pd.read_csv(url)
malecount = len(titanic[titanic['sex'] == 'male']['sex'])
femalecount = len(titanic[titanic['sex'] == 'female']['sex'])
# Data to plot a PIE
labels = ['Male', 'Female']
sizes = [malecount, femalecount]
colors = ['blue', 'pink']
# Pie Plot
plt.pie(sizes, explode=(0.01,0), labels=labels,colors=colors,autopct='%1.1f%%', shadow=True, startangle=90)
plt.axis('equal')
plt.show()
```

```python
# Data to plot a Scatter
colors = np.where(titanic.sex == 'male', 'blue', 'purple')
groups = np.where(titanic.sex == 'male', 'M', 'F')
plt.scatter(titanic.fare, titanic.age, c=colors, label='female')
#  ScatterPlot
plt.title('Scatter plot Fare vs Age')
plt.xlabel('Fare')
plt.ylabel('Age')
plt.legend()
plt.show()
```
